# 데이터 찾아보기

EDA 과제를 위해 다양한 공개 데이터셋을 활용할 수 있다.
아래 플랫폼들은 초보자부터 실무 수준까지 널리 사용되는 데이터 제공 사이트이다.

## 1. sklearn Datasets

* Python 라이브러리인 `scikit-learn`에서 기본 제공하는 데이터셋
* 별도 다운로드 없이 바로 사용 가능
* 입문 실습 및 알고리즘 테스트에 적합

```markdown
- 특징: 설치만 하면 바로 사용 가능, 구조가 단순함
- 추천 용도: 첫 EDA 연습, 모델 학습 테스트
- 링크: https://scikit-learn.org/stable/datasets.html
```

## 2. UCI Machine Learning Repository

* 오래된 전통의 머신러닝 데이터 저장소
* 학술/연구 기반 데이터가 많음

```markdown
- 특징: 정제된 데이터, 다양한 문제 유형 포함
- 단점: 일부 데이터는 설명이 부족하거나 오래됨
- 추천 용도: 이론 기반 실습, 다양한 유형 경험
- 링크: https://archive.ics.uci.edu/
```

## 3. Kaggle Datasets

* 가장 많이 사용되는 데이터 플랫폼
* 실무에 가까운 데이터 + 커뮤니티 + 코드 공유

```markdown
- 특징: 데이터 + 코드 + 설명 + 토론까지 제공
- 장점: 실무 감각 익히기 좋음
- 추천 용도: 프로젝트, 포트폴리오, 실전 연습
- 링크: https://www.kaggle.com/datasets
```

## 4. Google Dataset Search

* 구글이 제공하는 데이터 검색 엔진
* 다양한 사이트의 데이터를 한 번에 검색 가능

```markdown
- 특징: “데이터 구글”이라고 보면 됨
- 추천 용도: 특정 주제 데이터 찾을 때
- 링크: https://datasetsearch.research.google.com/
```

## 5. 공공데이터 포털 (Korea Public Data)

* 정부/공공기관 데이터 제공

```markdown
- 특징: 실제 정책/행정 데이터
- 추천 용도: 현실 문제 기반 분석
- 링크: https://www.data.go.kr/
```

## 6. GitHub (데이터 저장소)

* 많은 프로젝트에서 데이터셋을 함께 공개

```markdown
- 특징: 코드 + 데이터 함께 존재
- 추천 용도: 실무 구조 파악, 프로젝트 참고
- 검색: "dataset csv github"
```

```markdown
[데이터 선택 기준]

1. 컬럼 수: 5~30개 (너무 많으면 분석 어려움)
2. 데이터 수: 최소 500건 이상
3. 타겟 변수 존재 여부 확인 (예측 문제일 경우)
4. 결측치/범주형/수치형이 적절히 섞여 있는 데이터 추천
```

# 📝 EDA 레포트 작성 구조 (Markdown 기준 예시)
EDA(Exploratory Data Analysis, 탐색적 데이터 분석) 레포트는 데이터를 처음 접했을 때 **구조를 이해하고, 패턴을 발견하며, 문제점을 파악**하는 과정을 **체계적으로 문서화**하는 보고서이다.


#### 1. **프로젝트 개요**

* 분석 목적: 무엇을 알고자 하는가?
* 데이터 출처 및 설명: 어디서 수집했는가? 어떤 데이터인가?

```markdown
## 1. 프로젝트 개요
- 목적: 고객 이탈 예측을 위한 데이터 구조 파악 및 전처리 방향 설정
- 데이터 출처: Kaggle - Telco Customer Churn Dataset
- 데이터 구성: 총 21개 컬럼, 7043개 행
```


#### 2. **데이터 기본 정보 확인**

* 행/열 수
* 각 변수 이름, 타입, 결측치 여부 등

```markdown
## 2. 데이터 기본 정보
- 총 데이터 수: 7043개
- 컬럼 수: 21개
- 주요 컬럼: `gender`, `SeniorCitizen`, `MonthlyCharges`, `Churn` 등
```


#### 3. **기술 통계 및 데이터 요약**

* 수치형 변수: 평균, 중앙값, 표준편차, 최소/최대
* 범주형 변수: 각 범주의 빈도수 및 비율

```markdown
## 3. 기술 통계 요약
### 수치형 변수
| 컬럼명           | 평균    | 중앙값  | 표준편차 | 최솟값 | 최댓값 |
|------------------|--------|--------|---------|--------|--------|
| MonthlyCharges   | 64.76  | 70.35  | 30.09   | 18.25  | 118.75 |

### 범주형 변수
| 컬럼명    | 고유값 | 최빈값 | 분포                           |
|-----------|--------|--------|--------------------------------|
| gender    | 2개    | Male   | Male: 51%, Female: 49%         |
| Churn     | 2개    | No     | No: 73.5%, Yes: 26.5%          |
```

#### 4. **결측치 및 이상치 탐색**

* 각 컬럼의 결측치 수
* 이상치 탐색 방법: boxplot, IQR, 도수분포 등

```markdown
## 4. 결측치 및 이상치 탐색
- `TotalCharges`: 11건 결측치 발견 → 0일 사용 고객으로 판단되어 제거
- `MonthlyCharges`: 이상치 없음 (boxplot 기준)
```

---

#### 5. **변수 간 관계 분석**

* 시각화: 히스토그램, 상자 그림, 산점도, heatmap
* 상관분석: 수치형 변수 간 상관계수
* 그룹 비교: 생존/이탈 여부에 따른 요금 차이 등

```markdown
## 5. 변수 간 관계 분석
- `MonthlyCharges`와 `Churn` 관계: 이탈 고객의 평균 요금이 더 높음
- `tenure`와 `Churn` 관계: 체류 기간이 길수록 이탈 확률이 낮음
```

#### 6. **파생 변수 생성 및 전처리 제안**

* 범주형 그룹핑, 나이 구간화 등
* 로그변환, 정규화 여부

```markdown
## 6. 파생 변수 및 전처리
- `tenure` → 구간화(`Short`, `Medium`, `Long`)
- `MonthlyCharges` → 로그 변환
```

#### 7. **요약 및 인사이트 도출**

* 주요 특징 정리
* 모델링 방향성 제시

```markdown
## 7. 요약 및 인사이트
- 이탈 고객은 월 요금이 높고, 체류 기간이 짧은 경향
- `Contract` 유형(1년/2년)은 이탈률과 큰 연관 있음
- 향후 모델에서 주요 변수로 사용 가능: `Contract`, `MonthlyCharges`, `tenure`
```
