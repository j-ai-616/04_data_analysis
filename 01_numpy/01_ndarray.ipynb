# NumPy
- 수치 연산에 최적화 된 파이썬 라이브러리
- 대량의 숫자 데이터를 빠르고 효율적으로 처리하기 위해 사용

## ndarray (다차원 배열)
- 다차원 배열 타입으로 데이터를 효율적으로 처리
  - "같은 타입"의 데이터만 담음
  - 여러 차원으로 확장 가능 (1차원, 2차원, 3차원 등)
  - 한 번에 연산 가능(반복문 사용하는 list 대비 훨씬 빠름)

```python
# 관례적으로 np라는 별칭 사용
import numpy as np
np.__version__
```

```python
# python list를 ndarray로 생성
# 1차원
arr = np.array([1, 2, 3])
arr
```

```python
# 2차원: 1차원 배열 여러개가 묶여있는 것
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
# 각 차원의 길이가 정확하게 일치해야 한다.
# arr2 = np.array([[1, 2, 3], [4, 5]]) # ValueError
arr2
```

```python
# list와 ndarray(n-dimentional array)의 타입과 값 비교
print(type([1, 2, 3]), [1, 2, 3])
print(type(arr), arr)
```

## ndarray 속성
- shape : 배열의 형태
- ndim : 배열의 차원 수
- size : 전체 요소 수
- dtype : 요소의 자료형

## dtype
dtype은 배열 요소의 자료형을 의미한다. 자주 보게 되는 자료형은 아래와 같다.
- 정수형: int32, int64
- 실수형: float32, float64
- 불리언형: bool
- 문자열형: 문자열 또는 유니코드 문자열

```python
print(arr)
print(arr2)
print(arr.shape, arr2.shape) 
print(arr.ndim, arr2.ndim) # 차원의 개수
print(arr.size, arr2.size) # 배열의 총 원소 개수
print(arr.dtype, arr2.dtype) # 모두 동일한 타입의 데이터, 동일한 메모리 크기만 사용
```

```python
# 실수형 ndarray(배열) 생성
arr3 = np.array([
    [1.23, 3.45],
    [5.67, 7.89],
    [12.34, 23.45]
])
print(arr3)
print(arr3.shape)
print(arr3.ndim)
print(arr3.size)
print(arr3.dtype)
```

```python
# dtype 지정
arr4 = np.array([
    [1.23, 3.45],
    [5.67, 7.89],
    [12.34, 23.45]
], dtype=int)
print(arr4.dtype, arr4)

arr5 = np.array([1, 0, 1, 0], dtype=bool)
print(arr5.dtype, arr5)

# asrtype() : 생성 후 자료형 변환
arr6 = np.array([
    [1.23, 3.45],
    [5.67, 7.89],
    [12.34, 23.45]
])
new_arr = arr6.astype(int)  # dtype을 int로 변환, 원본 변경하지 않고 새로운 배열 반환
print(new_arr.dtype, new_arr)
print(arr6.dtype, arr6)  # 원본은 유지됨
```

```python
# ndarray는 하나의 타입만 요소로 가질 수 있다.
arr = np.array([1, 2, 3, 'hi'])
print(arr) # 내부적으로 타입 일치시켜 버림: 모든 요소를 문자열로 변환
print(arr.dtype)    # 유니코드 방식의 문자열 타입으로 변경 됨
```

## 메소드를 이용하여 특정 형식의 ndarray 생성하기
원하는 값으로 채워진 배열을 만들 수 있다.
- np.zeros(): 0으로 채운 배열 생성
- np.ones(): 1로 채운 배열 생성
- np.full(): 지정한 값으로 채운 배열 생성

```python
# np.zeros(shape: tuple))
arr1 = np.zeros((3, 2), dtype=float) # 3행 2열의 배열 생성, 모든 요소는 0, dtype=float가 기본 값
print(arr1)
arr2 = np.zeros([3, 2], dtype=int) # shape을 리스트로도 전달 가능, 보편적으로는 튜플로 전달
print(arr2)
```

```python
# np.ones(shape)
arr3 = np.ones((3, 2))
print(arr3)

# np.full(shape, n)
arr4 = np.full((4, 3), 3) # 4행 3열의 배열 생성, 모든 요소는 3
print(arr4)
```

- np.zeros_like(): 기존 배열과 같은 shape의 0 배열 생성
- np.ones_like(): 기존 배열과 같은 shape의 1 배열 생성
- np.full_like(): 기존 배열과 같은 shape의 지정 값 배열 생성

```python
data = np.array([[1, 2, 3], [4, 5, 6]])  # shape => (2, 3)

arr5 = np.zeros_like(data)
print(arr5)

arr6 = np.ones_like(data)
print(arr6)

arr7 = np.full_like(data, 7)
print(arr7)
```

## 특정 범위의 ndarray 생성하기
일정한 규칙을 가진 숫자 배열을 만들 수 있다.
- np.arrange(): 일정 간격의 값 생성
- np.linspace(): 구간을 동일한 개수로 나누어 생성
- np.logspace(): 로그 간격으로 값 생성

```python
# np.arange(start, stop, step)
arr1 = np.arange(0, 10)    # step=1 default
print(arr1)

arr2 = np.arange(0, 10, 2)
print(arr2)
```

```python
# np.linspace(start, stop, len) : start-end(포함)까지 len개의 균등간격 ndarray 반환
arr3 = np.linspace(0, 10, 10)
print(arr3)

arr4 = np.linspace(0, 10, 11)
print(arr4)
```

```python
# np.logspace(start, stop, len, base) : start부터 end(포함)까지 len개의 숫자를 지수로 하는 base 거듭제곱
arr5 = np.logspace(0, 10, 10, base=2)
print(arr5)

arr6 = np.logspace(0, 10, 10)   # base=10이 default
print(arr6)
# 1.e+01 : (가수부)e(지수부)
```
