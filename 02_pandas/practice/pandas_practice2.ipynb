## 0. 시작 데이터 만들기

실습 안내:
- 이 데이터는 이후 모든 문제에서 공통으로 사용한다.
- 날짜 컬럼은 미리 datetime으로 변환해 둔다.
- 총액은 수량 × 단가로 계산한다.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Mac 기준 한글 폰트 설정
plt.rcParams["font.family"] = "AppleGothic"
# 마이너스 부호 깨짐 문제 방지
plt.rcParams["axes.unicode_minus"] = False

orders = pd.DataFrame({
    '주문일자': ['2026-01-03', '2026-01-05', '2026-01-05', '2026-01-08', '2026-01-10',
             '2026-02-02', '2026-02-10', '2026-02-18', '2026-03-03', '2026-03-11'],
    '고객명': ['김철수', '이영희', '박민수', '최지영', '정다은', '홍길동', '강민호', '오세훈', '장미라', '서지수'],
    '지역': ['서울', '부산', '서울', '대전', '부산', '서울', '서울', '부산', '대전', '서울'],
    '카테고리': ['전자', '식품', '전자', '패션', '식품', '전자', '패션', '식품', '전자', '패션'],
    '상품명': ['노트북', '사과', '마우스', '자켓', '우유', '키보드', '운동화', '라면', '모니터', '가방'],
    '수량': [1, 10, 2, 1, 3, 1, 2, 5, 1, 1],
    '단가': [1500000, 3000, 25000, 120000, 2500, 50000, 90000, 4000, 350000, 70000]
})

orders['주문일자'] = pd.to_datetime(orders['주문일자'])
orders['총액'] = orders['수량'] * orders['단가']
orders
```

## 1. apply / map
1) 총액이 100000 이상이면 '고가', 아니면 '일반'을 반환하는 가격등급 열을 apply로 만드시오.
2) 지역 코드 맵을 사용해 지역코드 열을 map으로 만드시오. (서울:S, 부산:B, 대전:D)

```python
# 여기에 작성
# 1.
orders['가격등급'] = orders['총액'].apply(lambda x: '고가' if x >= 100000 else '일반')

# 2.
region_code_map = {
    '서울': 'S',
    '부산': 'B',
    '대전': 'D'
}

orders['지역코드'] = orders['지역'].map(region_code_map)

orders
```

## 2. replace / np.where
1) 카테고리의 '식품'을 '식음료'로 바꾸시오.
2) 총액이 100000 이상이면 VIP, 아니면 NORMAL인 고객등급 열을 np.where로 만드시오.

주의:
- 이후 집계 결과에는 변경된 카테고리명(식음료)이 반영된다.

```python
# 여기에 작성
# 1.
orders['카테고리'] = orders['카테고리'].replace('식품', '식음료')

# 2.
orders['고객등급'] = np.where(orders['총액'] >= 100000, 'VIP', 'NORMAL')

orders
```

## 3. np.select
총액 기준으로 주문등급 열을 만드시오.
- 500000 이상: A
- 100000 이상 500000 미만: B
- 그 외: C

```python
# 여기에 작성
conditions = [
    orders['총액'] >= 500000,
    (orders['총액'] >= 100000) & (orders['총액'] < 500000)
]

choices = ['A', 'B']

orders['주문등급'] = np.select(conditions, choices, default='C')

orders
```

## 4. concat
신규 주문 2건이 아래와 같이 들어왔다.
new_orders를 orders와 세로로 연결한 result_concat을 만드시오.

주의:
- 새 주문 데이터에는 아직 일부 파생 컬럼이 없을 수 있다.
- 필요하면 orders의 컬럼 구조에 맞춰 정리한 뒤 연결해도 된다.

```python
new_orders = pd.DataFrame({
    '주문일자': pd.to_datetime(['2026-03-20', '2026-03-25']),
    '고객명': ['한지민', '이도윤'],
    '지역': ['부산', '서울'],
    '카테고리': ['전자', '식음료'],
    '상품명': ['이어폰', '커피'],
    '수량': [2, 4],
    '단가': [80000, 5000],
    '총액': [160000, 20000]
})

# 여기에 작성
result_concat = pd.concat([orders, new_orders], ignore_index=True, sort=False)
result_concat
```

## 5. merge
고객 등급표를 결합하시오.
고객명 기준으로 left join 하여 orders_grade를 만드시오.

확인 포인트:
- 등급표에 없는 고객은 회원등급이 NaN으로 남는다.
- left join이므로 orders의 행 수는 유지된다.

```python
grade_df = pd.DataFrame({
    '고객명': ['김철수', '이영희', '홍길동', '강민호'],
    '회원등급': ['Gold', 'Silver', 'Gold', 'Bronze']
})

# 여기에 작성
orders_grade = pd.merge(orders, grade_df, on='고객명', how='left')
orders_grade
```

## 6. pivot_table
지역별, 카테고리별 총액 합계를 피벗 테이블로 만드시오.

참고:
- 집계 결과에 따라 값이 없는 조합은 NaN으로 보일 수 있다.

```python
# 여기에 작성
pivot_result = orders.pivot_table(
    index='지역',
    columns='카테고리',
    values='총액',
    aggfunc='sum'
)

pivot_result
```

## 7. melt
아래 wide_df를 long 형태로 바꾸시오. 열 이름은 월, 매출을 사용하시오.

```python
wide_df = pd.DataFrame({
    '상품명': ['노트북', '마우스'],
    '1월': [100, 50],
    '2월': [120, 60],
    '3월': [110, 70]
})

# 여기에 작성
long_df = pd.melt(
    wide_df,
    id_vars='상품명',
    var_name='월',
    value_name='매출'
)

long_df
```

## 8. groupby 기본
1) 카테고리별 총액 합계를 구하시오.
2) 지역별 평균 총액을 구하시오.

가능하면 결과를 각각 다른 변수에 저장해보시오.

```python
# 여기에 작성
# 1.
category_total = orders.groupby('카테고리')['총액'].sum()

# 2.
region_avg = orders.groupby('지역')['총액'].mean()

print(category_total)
print(region_avg)
```

## 9. agg / named aggregation
카테고리별로 아래 값을 한 번에 구하시오.
- 주문수
- 평균총액
- 최고총액

```python
# 여기에 작성
category_summary = orders.groupby('카테고리').agg(
    주문수=('총액', 'count'),
    평균총액=('총액', 'mean'),
    최고총액=('총액', 'max')
)

category_summary
```

## 10. transform
카테고리 평균 총액을 각 행에 붙이고, 평균대비차이 열을 만드시오.

```python
# 여기에 작성
# 카테고리별 평균 총액을 각 행에 붙이기
orders['카테고리평균총액'] = orders.groupby('카테고리')['총액'].transform('mean')
# 각 행의 총액 - 카테고리 평균총액
orders['평균대비차이'] = orders['총액'] - orders['카테고리평균총액']

orders
```

## 11. 날짜 처리
1) 주문월 열을 만드시오.
2) 주문요일명 열을 만드시오.

참고:
- `day_name()`은 기본적으로 영어 요일명을 반환한다.

```python
# 여기에 작성
# 1.
orders['주문월'] = orders['주문일자'].dt.month
# 2.
orders['주문요일명'] = orders['주문일자'].dt.day_name()

orders
```

## 12. 시계열 인덱스와 리샘플링
1) 주문일자를 인덱스로 설정한 ts_orders를 만드시오.
2) 월말 기준 월별 총액 합계를 구하시오.

```python
# 여기에 작성
# 1.
ts_orders = orders.set_index('주문일자')
# 2.
monthly_total = ts_orders['총액'].resample('ME').sum()

print(ts_orders)
print('---')
print(monthly_total)
```

## 13. rolling
일자별 총액 합계를 구한 뒤 3일 이동평균을 계산하시오.

참고:
- 일 단위로 리샘플링하면 주문이 없는 날짜도 포함될 수 있다.

```python
# 여기에 작성
# 일자별 총액 합계
daily_total = ts_orders['총액'].resample('D').sum()
# 3일 이동평균
rolling_3day = daily_total.rolling(3).mean()

print(daily_total)
print('---')
print(rolling_3day)
```

## 14. shift / diff / pct_change
일자별 총액 합계를 기준으로
- 전일총액
- 전일대비
- 변화율
열을 만드시오.

참고:
- 전일 값이 0인 경우 변화율 해석에 주의한다.

```python
# 여기에 작성
shift_result = pd.DataFrame({
    '일자별총액': daily_total
})

# 전일총액
shift_result['전일총액'] = shift_result['일자별총액'].shift(1)

# 전일대비
shift_result['전일대비'] = shift_result['일자별총액'].diff()

# 변화율
shift_result['변화율'] = shift_result['일자별총액'].pct_change()

shift_result
```

## 15. 시각화
1) 월별 총액 합계를 막대 그래프로 그리시오.
2) 일자별 총액과 3일 이동평균을 선 그래프로 그리시오.

```python
# 여기에 작성
# 1.
monthly_total.plot(kind='bar', title='월별 총액 합계')
plt.show()
# 2.
plt.figure(figsize=(10, 4))
plt.plot(daily_total.index, daily_total.values, label='일자별 총액')
plt.plot(rolling_3day.index, rolling_3day.values, label='3일 이동평균')
plt.title('일자별 총액과 3일 이동평균')
plt.legend()
plt.show()
```
